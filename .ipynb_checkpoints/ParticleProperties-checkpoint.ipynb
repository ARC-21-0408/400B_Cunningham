```python
import numpy as np
import astropy.units as u
import import_ipynb
from ReadFile import Read

def ParticleInfo(filename,wow,partnum):
    a,b,c=Read(filename)
    
    x=0
    y=0
    z=0
    index=np.where(c['type']==wow)
    x=c['x'][index]*u.kpc
    y=c['y'][index]*u.kpc
    z=c['z'][index]*u.kpc
    xcom=x[partnum]
    ycom=y[partnum]
    zcom=z[partnum]
    r=0
    r=np.around((xcom**2+ycom**2+zcom**2)**.5,3)
    
    vx=0
    vy=0
    vz=0
    vx=c['vx'][index]*u.km/u.s
    vy=c['vy'][index]*u.km/u.s
    vz=c['vz'][index]*u.km/u.s
    vxcom=vx[partnum]
    vycom=vy[partnum]
    vzcom=vz[partnum]
    vel=0
    vel=np.around((vxcom**2+vycom**2+vzcom**2)**.5,3)
    
    m=c['m'][index]*10**10*u.solMass
    mass=m[partnum]
    
    return r,vel,mass

e=0
f=0
g=0
namefile='MW_000.txt'
wee=2
numpart=99
i=2 #index 100th disk particle is the 99 item in the array becuase the array starts at 0
e,f,g=ParticleInfo(namefile,wee,numpart)
print(e,f,g)
print(np.around(e.to(u.lyr),3))
```
